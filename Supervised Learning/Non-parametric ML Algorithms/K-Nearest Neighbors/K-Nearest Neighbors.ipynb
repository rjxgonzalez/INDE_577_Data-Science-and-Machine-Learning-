```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme()

data = pd.read_csv("Clothing Size.csv") # Reading in the Gender Classification dataset
```

```python
data
```

```python
small_w = data.iloc[:100]["Weight (kg)"] # storing the width feature measurement of the first 100 rows (females)
small_h = data.iloc[:100]["Height (cm)"] # storing the height feature measurement of the first 100 rows (females)

medium_w = data.iloc[3357:3457]["Weight (kg)"]    # storing the width feature measurement of the last 100 rows (males)
medium_h= data.iloc[3357:3457]["Height (cm)"]   # storing the height feature measurement of the last 100 rows (males)

large_w = data.iloc[7255:7355]["Weight (kg)"]
large_h = data.iloc[7255:7355]["Height (cm)"]
```

```python
plt.figure(figsize = (15, 10))
plt.scatter(small_w,
            small_h,
            color = "lightseagreen",
            label = "Small")
plt.scatter(medium_w,
            medium_h,
            color = "magenta",
            label = "Medium")
plt.scatter(large_w,
            large_h,
            color = "red",
            label = "Large")         
plt.xlabel("Weight (kg)", fontsize = 15)
plt.ylabel("Height (cm)", fontsize = 15)
plt.legend(fontsize = 12)
plt.title("Clothing Size", fontsize = 18)
plt.show()
```
