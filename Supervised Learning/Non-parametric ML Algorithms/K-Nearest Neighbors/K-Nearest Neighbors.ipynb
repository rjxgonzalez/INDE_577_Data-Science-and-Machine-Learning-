```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme()

data = pd.read_csv("Gender Classification.csv") # Reading in the Gender Classification dataset

# Plot the data
genders = ["Female", "Male"]
colors = ["red", "lightseagreen"]

fig, ax = plt.subplots(figsize = (10, 8))
for genders, color in zip(genders, colors):
    temp_df = data[data.Gender == genders]
    ax.scatter(temp_df.iloc[:100]["Forehead Width (cm)"] ,
               temp_df.iloc[:100]["Forehead Height (cm)"] ,
               c = color,
               label = genders, 
               )
    
ax.set_xlabel("sepal length [cm]", fontsize = 15)
ax.set_ylabel("sepal width [cm]", fontsize = 15)
ax.legend()
plt.show()
```
